```python
def read_input(infile):
    rows = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            rows.append([int(v) for v in line.strip().split()])
    return rows

def calc_checksum(rows):
    s = 0

    for r in rows:
        s += max(r) - min(r)

    return s

def calc_div_checksum(rows):
    s = 0

    for r in rows:
        row_done = False
        for i, v1 in enumerate(r):
            for v2 in r[i+1:]:
                if v1 % v2 == 0:
                    s += v1 / v2
                    row_done = True
                    break
                if v2 % v1 == 0:
                    s += v2 / v1
                    row_done = True
                    break
            if row_done:
                break

    return s
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = calc_checksum(read_input('input02a.txt'))

print(f'Checksum is {res}')

assert res == 18

print('\nPuzzle case\n-----------\n')

res = calc_checksum(read_input('input02.txt'))

print(f'Checksum is {res}')

assert res == 36174

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = calc_div_checksum(read_input('input02b.txt'))

print(f'Checksum is {res}')

assert res == 9

print('\nPuzzle case\n-----------\n')

res = calc_div_checksum(read_input('input02.txt'))

print(f'Checksum is {res}')

assert res == 244
```
