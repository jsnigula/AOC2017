```python
def read_input(infile):
    passphrases = []
    with open(infile, 'r') as inf:
        for line in inf.readlines():
            passphrases.append(line.strip())
    return passphrases

def is_valid(passphrase):
    tokens = [s for s in passphrase.split()]
    if len(set(tokens)) == len(tokens):
        return True

    return False

def is_valid_with_added_security(passphrase):
    tokens = [s for s in passphrase.split()]
    token_set = list(set(tokens))
    if len(token_set) == len(tokens):
        for i, t1 in enumerate(tokens[:-1]):
            s_t1 = sorted([c for c in t1])
            for t2 in tokens[i+1:]:
                s_t2 = sorted([c for c in t2])
                if s_t2 == s_t1:
                    return False
        return True

    return False

def count_valid_passphrases(passphrases):
    n_valid = 0
    for p in passphrases:
        if is_valid(p):
            n_valid += 1

    return n_valid

def count_valid_passphrases_with_added_security(passphrases):
    n_valid = 0
    for p in passphrases:
        if is_valid_with_added_security(p):
            n_valid += 1

    return n_valid
        
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test cases\n----------\n')

assert is_valid('aa bb cc dd ee') == True
assert is_valid('aa bb cc dd aa') == False
assert is_valid('aa bb cc dd aaa') == True

print('Test cases passed')

print('\nPuzzle case\n-----------\n')

res = count_valid_passphrases(read_input('input04.txt'))

print(f'Number of valid passphrases is {res}')

assert res == 477

print('\n*******\nPuzzle2\n*******\n')

print('Test cases\n----------\n')

assert is_valid_with_added_security('abcde fghij') == True
assert is_valid_with_added_security('abcde xyz ecdab') == False
assert is_valid_with_added_security('a ab abc abd abf abj') == True
assert is_valid_with_added_security('iiii oiii ooii oooi oooo') == True
assert is_valid_with_added_security('oiii ioii iioi iiio') == False

print('Test cases passed')

print('\nPuzzle case\n-----------\n')

res = count_valid_passphrases_with_added_security(read_input('input04.txt'))

print(f'Number of valid passphrases is {res}')

assert res == 167
```
