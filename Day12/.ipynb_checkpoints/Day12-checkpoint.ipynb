```python
from collections import defaultdict
```

```python
def read_input(infile):

    pipes = defaultdict(set)

    with open(infile, 'r') as inf:
        for line in inf.readlines():
            p1, p = line.strip().split(' <-> ')
            for p2 in p.split(','):
                pipes[int(p1)].add(int(p2))
                pipes[int(p2)].add(int(p1))
    return pipes

def get_group(i, pipes, checked):

    to_check = [i]
    cons = set()

    while len(to_check) > 0:
        n = to_check.pop(0)
        checked.append(n)
        for p in pipes[n]:
            cons.add(p)
            if p not in to_check and p not in checked:
                to_check.append(p)
    return cons

def connections(pipes):

    checked = []
    cons = get_group(0, pipes, checked)

    return len(cons)

def groups(pipes):

    checked = []
    n_groups = 0
    for p in pipes:
        if p not in checked:
            n_groups += 1
            get_group(p, pipes, checked)
    return n_groups
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = connections(read_input('input12a.txt'))

print(f'Number of connections is {res}')

assert res == 6

print('\nPuzzle case\n-----------\n')

res = connections(read_input('input12.txt'))

print(f'Number of connections is {res}')

assert res == 128

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = groups(read_input('input12a.txt'))

print(f'Number of groups is {res}')

assert res == 2

print('\nPuzzle case\n-----------\n')

res = groups(read_input('input12.txt'))

print(f'Number of groups is {res}')

assert res == 209
```
