```python
def read_input(infile):
    with open(infile, 'r') as inf:
        return (int(v) for v in inf.readline().strip().split(','))
```

```python
def calc_value_a(start):
    return (start * 16807) % 2147483647

def calc_value_b(start):
    return (start * 48271) % 2147483647

def calc_value_a2(start):
    while True:
        v = (start * 16807) % 2147483647
        if v % 4 == 0:
            return v
        start = v

def calc_value_b2(start):
    while True:
        v = (start * 48271) % 2147483647
        if v % 8 == 0:
            return v
        start = v

def test(a, b):
    # known = [start]
    v_a = a
    v_b = b
    last = 0
    n_matches = 0
    for i in range(40000000):
        v_a = calc_value_a(v_a)
        v_b = calc_value_b(v_b)
        # print(v_a, v_b)
        # if f'{v_a:032b}'[-16:] == f'{v_b:032b}'[-16:]:
        if (v_a & 0xffff) == (v_b & 0xffff):
            n_matches += 1
            # print(i-last)
            last = i
        # known.append(v)

    return n_matches


def test2(a, b):
    # known = [start]
    v_a = a
    v_b = b
    last = 0
    n_matches = 0
    for i in range(5000000):
        v_a = calc_value_a2(v_a)
        v_b = calc_value_b2(v_b)
        # print(v_a, v_b)
        # if f'{v_a:032b}'[-16:] == f'{v_b:032b}'[-16:]:
        if (v_a & 0xffff) == (v_b & 0xffff):
            n_matches += 1
            # print(i-last)
            last = i
        # known.append(v)

    return n_matches
```

```python
print('*******\nPuzzle1\n*******\n')

print('Test case\n---------\n')

res = test(*read_input('input15a.txt'))

print(f'Number of matches is {res}')

assert res == 588

print('\nPuzzle case\n-----------\n')

res = test(*read_input('input15.txt'))

print(f'Number of matches is {res}')

assert res == 650

print('\n*******\nPuzzle2\n*******\n')

print('Test case\n---------\n')

res = test2(*read_input('input15a.txt'))

print(f'Number of matches is {res}')

assert res == 309

print('\nPuzzle case\n-----------\n')

res = test2(*read_input('input15.txt'))

print(f'Number of matches is {res}')

assert res == 336
```
